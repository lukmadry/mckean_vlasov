```python
def read_cpp_data(fname):
    with open(fname, 'r') as f:
        data = f.readlines()
        data = [float(y) for y in data[0].split()]
    return data
```

```python
def read_cpp_data_twodim(fname):
    with open(fname, 'r') as f:
        data = f.readlines()
        data = [[float(y) for y in d.split()] for d in data]
    return data
```

```python
gaus = read_cpp_data('gaussians.txt')
plt.hist(gaus, bins=28)
```
